## Linear model with one numerical variable

Execute the following cell to load the [ISLR2](https://cran.rstudio.com/web/packages/ISLR2/index.html) and [Tidverse](https://www.tidyverse.org/) packages.

````r
library('tidyverse')
library('ISLR2')
````

The `ISLR2` package provides a dataset called `Boston` that we will use in this assignment.

````r
Boston |> glimpse()
````

üö® Use the **Jupyter Lab Contextual Help Feature** to see the documentation for this data. 

‚úîÔ∏è**In the markdown cell below, add definitions for the `lstat` and `medv` variables in the `Boston` data.** 

- `lstat`: lower status of the population (percent) - 1 of 13 variables from the `Boston` dataset.
- `medv`: median value of owner-occupied homes in $1000s - another of the 13 variables from the `Boston` dataset.

Let's plot `medv` (y-axis) versus `lstat`.

````r
p = ggplot(Boston, aes(x = lstat, y = medv)) + geom_point()

p
````

‚úîÔ∏èDoes the relationship appear to be positive or negative? Does it look to be reasonably linear?

**Answer:**

The relationship appears to be negative, because as the `medv` values get smaller, the `lstat` values get larger.

I would expect to see a negative slope when we look at the regression table for the model.

The relationship does look _reasonably_ linear, though I see some curvature, concave upward, so maybe a curve would be a good fit, too.

### Let's build a model!

We need to first load the [Tidymodels]() package.

````r
library('tidymodels')
````

First we specify our model as `linear regression` using the `lm` engine.

````r
# This step produces the *mod* object to store the specification/instructions for generating a linear regression model.
# This same *mod* object can be used throughout this exercise each time we want to model something new.

mod = linear_reg() |> set_engine("lm")

mod
````

Next we "fit" our model by supplying the `formula` and the data.

````r
# The fit function itself takes a function as its first argument that designates the predicted variable as a function of the predictor variable.

mod_fit = mod |> fit(medv ~ lstat, data = Boston)

mod_fit
````

### Get the regression table

````r
# We can use the tidy function to get a table of our model information
tidy(mod_fit)
````

‚úîÔ∏èComparing on the chart above and the regression table, does the **sign** of the `estimate` for the `lstat` term coefficient fit your expectations?

**Answer:**

Yes, it does. I expected a negative slope value, i.e., the `estimate` for the `lstat` term coefficient. This matched my expectations, which is a good learning reinforcement.

### Making predictions and visualizing the model

We can use the `augment` function to "predict" `medv` for all the values in our original dataset. We will capture these predictions in a new data frame called `Boston2`. The predicted values are found in the `.pred` column.

````r
# The augment function will generate predicted values according to the regression model and add two new columns:
# The .pred column will have the predicted values for medv.
# The .resid column will have values that are the diffs between the actual and predicted medv values in this dataset (medv - .pred = .resid)

Boston2 = augment(mod_fit, Boston)

glimpse(Boston2)
````

Let's visualize our model.

````r
p = ggplot(Boston2, aes(x = lstat)) +
    geom_point(aes(y = medv)) + 
    geom_line(aes(y = .pred), color = 'coral', linewidth = 1.5)

p
````

## Linear model with one categorical variable

We will use the `Carseats` data from the `ISLR2` package for the following exercise.

````r
Carseats |> glimpse()
````

Below is a plot of `sales` versus `ShelveLoc`.

````r
# I wanted to order the swarms by Bad, Medium, Good, and I checked the documentation for a way to do that.

Carseats$ShelveLoc = factor(Carseats$ShelveLoc, levels = c("Bad", "Medium", "Good"))

p = ggplot(Carseats, aes(x = ShelveLoc, y = Sales)) + 
    geom_point(position = position_jitter(width = 0.2, height = 0)) # Using 0.2 width to match the plot below.

p
````

‚úîÔ∏èDoes it look like a "Good" shelf location is associated with more car seat sales?

**Answer:**

It looks like a "Good" shelf location **is associated** with more car seat sales.

In cells below, using the `Carseats` data build a model of `Sales` with the `ShelveLoc` variable as a predictor.

**Include the following:**
- Show a regression table of your model parameters.
- Visualize the model with `ggplot2`.

````r
# First build a model, reusing *mod* and selecting our predictor and predicted variables.

mod_fit_shelveloc = mod |> fit(Sales ~ ShelveLoc, data = Carseats)

# Then show a regression table for the model parameters.

tidy(mod_fit_shelveloc)
````

````r
# Using augment to generate predicted values and a new dataframe.

Carseats2 = augment(mod_fit_shelveloc, Carseats)

# Let's see what that looks like.

Carseats2 |> glimpse()
````
